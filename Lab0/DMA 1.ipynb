```python
wage = pd.read_csv("wages_hours.csv", sep = "\t")
wage.head()
```

```python
wage1 = wage [["AGE","RATE"]]
wage1.head()
wage1.plot()
plt.show()
```

```python
age_sort = wage1.sort_values(["AGE"])
age_sort.set_index("AGE",inplace = True)
```

```python
age_sort.plot()
plt.show()
```

```python
web = pd.read_csv("visitors.csv", skiprows=4,names = ["DATE","VISITORS"])
```

```python
web.head()
```

```python
web_new = pd.read_csv("visitors-new.csv", skiprows=4,names = ["DATE","VISITORS-NEW"])
web_new.head()
```

```python
web_comb = pd.merge(web,web_new)
web_comb.head()
```

```python
web_sort = web_comb.sort_values(["DATE"])
web_sort.set_index("DATE",inplace = True)
```

```python
web_sort.plot()
plt.show()
```
